```python
#from rpunct import RestorePuncts
from wordsegment import load, segment
from punctfix import PunctFixer
load()

def fix(data):
  data=" ".join(segment(data))
  fixer = PunctFixer()
  #data=PunctFixer().fix(data)
  data=fixer.punctuate(data)
  return data
```

```python

with open("text.txt", "r") as file:
    content = file.read()

fix(content)
```
